# Exploratory Analysis – Energy Consumption

This notebook provides a quick exploratory analysis of the smart meter electricity dataset.

**What you'll find here**
- Dataset overview (shape, columns, time coverage)
- Missing value inspection
- Descriptive statistics
- Time-series visualization
- Consumption distribution

```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

from src.data_loader import DataLoaderConfig, load_and_clean_data

DATA_PATH = Path('..') / 'data' / 'household_power_consumption.csv'
cfg = DataLoaderConfig(csv_path=DATA_PATH)
df = load_and_clean_data(cfg)
df.head()
```

```python
# Dataset overview
df.shape, df.columns
```

```python
# Time coverage
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
df['datetime'].min(), df['datetime'].max()
```

```python
# Missing value inspection
df.isna().mean().sort_values(ascending=False)
```

```python
# Descriptive statistics
df.describe(include='all')
```

```python
# Time-series visualization (hourly resample for readability)
df_ts = df.set_index('datetime').sort_index().resample('h').mean(numeric_only=True)
ax = df_ts['global_active_power'].plot(figsize=(12, 4), linewidth=1, title='Global Active Power (Hourly Mean)')
ax.set_xlabel('Time')
ax.set_ylabel('Global Active Power')
plt.show()
```

```python
# Consumption distribution
plt.figure(figsize=(10, 4))
plt.hist(df_ts['global_active_power'].dropna(), bins=80)
plt.title('Distribution of Global Active Power (Hourly Mean)')
plt.xlabel('Global Active Power')
plt.ylabel('Count')
plt.show()
```
